```python
# ===================== INSTALL =====================
!pip install -q inference-sdk deepface opencv-python-headless pandas scikit-learn pytz

# ===================== IMPORTS =====================
import os, cv2, json, sqlite3, pickle
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
from sklearn.cluster import KMeans
from sklearn.neighbors import KNeighborsClassifier
from deepface import DeepFace
import pytz
from google.colab import drive
from inference_sdk import InferenceHTTPClient

# ===================== MOUNT DRIVE =====================
drive.mount("/content/drive")

# ===================== ROBOFLOW CLIENT =====================
CLIENT = InferenceHTTPClient(
    api_url="https://serverless.roboflow.com",
    api_key="p5peDvSzkBP2UVS4cCNa"     # <-- CHANGE
)
ROBOFLOW_MODEL_ID = "nv2-fruy1/3"

# ===================== PATHS =====================
ENROLL_DIR  = "/content/drive/MyDrive/Photo"
VIDEO_PATH  = "/content/drive/MyDrive/Video/pdv_rc_ss1.mp4"
REPORT_PATH = "/content/drive/MyDrive/attendance_report.csv"
DB_PATH     = "attendance.db"

# ===================== PARAMETERS =====================
FRAME_SKIP = 25
MAX_FRAMES = 1200
KMEANS_CLUSTERS = 2
KNN_K = 3
TIME_CUTOFF_MIN = 8

ROBOFLOW_CONF_MIN = 0.45
FUSED_THRESHOLD  = 0.80
MIN_MARGIN       = 0.12   # OPEN-SET REJECTION
VOTE_THRESHOLD = 3   # must appear in 3 different frames

# ===================== DATABASE =====================
def init_db():
    conn = sqlite3.connect(DB_PATH)
    c = conn.cursor()
    c.execute("CREATE TABLE users(id INTEGER PRIMARY KEY, name TEXT, threshold REAL DEFAULT 0.6)")
    c.execute("CREATE TABLE embeddings(user_id INTEGER, emb TEXT)")
    c.execute("CREATE TABLE centroids(user_id INTEGER, emb TEXT)")
    c.execute("CREATE TABLE attendance(user_id INTEGER, ts TEXT)")
    conn.commit()
    conn.close()

# ===================== ROBOFLOW FACE DETECTOR =====================
def detect_faces_roboflow(frame):
    result = CLIENT.infer(frame, model_id=ROBOFLOW_MODEL_ID)
    preds = result.get("predictions", [])
    print(f"[DEBUG] Roboflow detections: {len(preds)}")

    boxes = []
    for p in preds:
        conf = p.get("confidence", 0)
        print("[DEBUG] RF conf:", conf)
        if conf < ROBOFLOW_CONF_MIN:
            continue

        x = int(p["x"] - p["width"] / 2)
        y = int(p["y"] - p["height"] / 2)
        w = int(p["width"])
        h = int(p["height"])
        boxes.append((x,y,w,h))

    return boxes

# ===================== ENROLLMENT =====================
def enroll():
    conn = sqlite3.connect(DB_PATH)
    c = conn.cursor()

    for name in sorted(os.listdir(ENROLL_DIR)):
        folder = os.path.join(ENROLL_DIR, name)
        if not os.path.isdir(folder):
            continue

        c.execute("INSERT INTO users(name) VALUES(?)", (name,))
        uid = c.lastrowid

        for imgf in os.listdir(folder):
            img = cv2.imread(os.path.join(folder, imgf))
            if img is None:
                continue

            # --- DETECT FACE USING SAME PIPELINE ---
            boxes = detect_faces_roboflow(img)
            if len(boxes) == 0:
                continue

            x,y,w,h = boxes[0]

            # --- SAME EXPANSION ---
            # --- RECENTER + SCALE FACE BOX (CRITICAL FIX) ---
            cx = x + w // 2
            cy = y + h // 2

            size = int(1.35 * max(w, h))   # head-sized box
            x1 = max(0, cx - size // 2)
            y1 = max(0, cy - size // 2)
            x2 = min(img.shape[1], cx + size // 2)
            y2 = min(img.shape[0], cy + size // 2)

            face = img[y1:y2, x1:x2]
            if face.size == 0:
                continue

            face = cv2.resize(face, (112,112))
            face = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)


            emb = DeepFace.represent(
                face,
                model_name="ArcFace",
                detector_backend="skip",   # IMPORTANT
                align=False,
                enforce_detection=False
            )[0]["embedding"]

            emb = np.array(emb, dtype=np.float32)
            emb /= np.linalg.norm(emb)

            c.execute(
                "INSERT INTO embeddings VALUES (?,?)",
                (uid, json.dumps(emb.tolist()))
            )

    conn.commit()
    conn.close()
    print("[OK] Enrollment completed")


# ===================== BUILD MODELS =====================
def build_models():
    conn = sqlite3.connect(DB_PATH)
    df = pd.read_sql("SELECT * FROM embeddings", conn)
    conn.close()

    X = np.array([json.loads(e) for e in df.emb])
    y = df.user_id.values

    conn = sqlite3.connect(DB_PATH)
    c = conn.cursor()

    for uid in np.unique(y):
        embs = X[y == uid]
        if len(embs) <= KMEANS_CLUSTERS:
            centers = [embs.mean(axis=0)]
        else:
            km = KMeans(n_clusters=KMEANS_CLUSTERS, random_state=0).fit(embs)
            centers = km.cluster_centers_

        for ctr in centers:
            ctr /= np.linalg.norm(ctr)
            c.execute("INSERT INTO centroids VALUES (?,?)",
                      (int(uid), json.dumps(ctr.tolist())))

    conn.commit()
    conn.close()

    knn = KNeighborsClassifier(n_neighbors=min(KNN_K, len(X)), metric="cosine")
    knn.fit(X, y)
    pickle.dump(knn, open("knn.pkl","wb"))

    print("[OK] Centroids + KNN ready")


#visualization embeedings

from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

def visualize_clusters():
    conn = sqlite3.connect(DB_PATH)
    df = pd.read_sql("SELECT * FROM embeddings", conn)
    users = pd.read_sql("SELECT * FROM users", conn)
    conn.close()

    X = np.array([json.loads(e) for e in df.emb])
    y = df.user_id.values
    id2name = dict(zip(users.id, users.name))

    pca = PCA(n_components=2)
    X2 = pca.fit_transform(X)

    plt.figure(figsize=(8,6))
    for uid in np.unique(y):
        idx = y == uid
        plt.scatter(X2[idx,0], X2[idx,1], label=id2name[uid])

    plt.legend()
    plt.title("Enrollment Embedding Clusters (ArcFace)")
    plt.show()


# ===================== ATTENDANCE =====================
def mark_attendance(uid):
    tz = pytz.timezone("Asia/Kolkata")
    now = datetime.now(tz)

    conn = sqlite3.connect(DB_PATH)
    c = conn.cursor()
    c.execute("SELECT ts FROM attendance WHERE user_id=? ORDER BY ts DESC LIMIT 1", (uid,))
    row = c.fetchone()

    if row:
        last = datetime.fromisoformat(row[0])
        if now - last < timedelta(minutes=TIME_CUTOFF_MIN):
            conn.close()
            return

    c.execute("INSERT INTO attendance VALUES (?,?)", (uid, now.isoformat()))
    conn.commit()
    conn.close()
    print("[ATTENDANCE]", uid)


from collections import defaultdict
vote_buffer = defaultdict(int)

# ===================== RECOGNITION =====================
def recognize():
    knn = pickle.load(open("knn.pkl","rb"))

    conn = sqlite3.connect(DB_PATH)
    cent = pd.read_sql("SELECT * FROM centroids", conn)
    users = pd.read_sql("SELECT * FROM users", conn)
    conn.close()

    C = np.array([json.loads(e) for e in cent.emb])
    U = cent.user_id.astype(int).values
    id2name = dict(zip(users.id.astype(int), users.name))

    marked_users = set()

    cap = cv2.VideoCapture(VIDEO_PATH)
    f = 0

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret or f > MAX_FRAMES:
            break

        f += 1
        if f % FRAME_SKIP != 0:
            continue

        for (x,y,w,h) in detect_faces_roboflow(frame):
            # ---- MINIMUM FACE SIZE CHECK (MANDATORY) ----
            if w < 45 or h < 45:
                continue
             # -------- EXPAND FACE BOX (CRITICAL) --------
            # --- RECENTER + SCALE FACE BOX (CRITICAL FIX) ---
            cx = x + w // 2
            cy = y + h // 2

            size = int(1.35 * max(w, h))   # head-sized box
            x1 = max(0, cx - size // 2)
            y1 = max(0, cy - size // 2)
            x2 = min(frame.shape[1], cx + size // 2)
            y2 = min(frame.shape[0], cy + size // 2)


            face = frame[y1:y2, x1:x2]
            if face.size == 0:
                continue

            face = cv2.resize(face, (112,112))
            face = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)


            emb = DeepFace.represent(
                face,
                model_name="ArcFace",
                detector_backend="skip",
                align=False,
                enforce_detection=False
            )[0]["embedding"]

            emb = np.array(emb, dtype=np.float32)
            emb /= np.linalg.norm(emb)

            dists = np.linalg.norm(C - emb, axis=1)
            sims  = C @ emb

            order = np.argsort(dists)
            best, second = order[0], order[1]

            best_dist = float(dists[best])
            second_dist = float(dists[second])
            margin = second_dist - best_dist

            best_uid = int(U[best])
            best_name = id2name[best_uid]
            best_sim = float(sims[best]) # Correctly re-added

            cv2.imwrite(f"/content/drive/MyDrive/debug_face_{best_name}_{f}.jpg",cv2.cvtColor(face, cv2.COLOR_RGB2BGR))# Correctly placed after best_name is defined

            user_thr = float(users.loc[users.id == best_uid, "threshold"].values[0])

            fused = 0.5 * (1 - best_dist/1.4) + 0.5 * best_sim
            knn_uid = int(knn.predict([emb])[0])

            STRONG = best_dist < user_thr * 0.75
            WEAK   = best_dist < user_thr
            FUSED_OK = fused >= FUSED_THRESHOLD
            MARGIN_OK = margin >= MIN_MARGIN

            accept = False
            if STRONG and FUSED_OK and MARGIN_OK:
                accept = True
            elif WEAK and FUSED_OK and MARGIN_OK and knn_uid == best_uid:
                accept = True

            print(
                f"[TRACE] {best_name} | "
                f"dist={best_dist:.3f} thr={user_thr:.3f} "
                f"sim={best_sim:.3f} fused={fused:.3f} "
                f"margin={margin:.3f} knn={knn_uid} accept={accept}"
            )

            if accept:
                vote_buffer[best_uid] += 1
                print(f"[VOTE] {best_name} -> {vote_buffer[best_uid]}")

                if vote_buffer[best_uid] >= VOTE_THRESHOLD and best_uid not in marked_users:
                    print("[CONFIRMED MATCH]", best_name)
                    mark_attendance(best_uid)
                    marked_users.add(best_uid) # Correctly re-added


    cap.release()

# ===================== EXPORT =====================
def export_csv():
    conn = sqlite3.connect(DB_PATH)
    df = pd.read_sql("""
        SELECT users.name, attendance.ts
        FROM attendance JOIN users
        ON attendance.user_id = users.id
    """, conn)
    conn.close()

    df.to_csv(REPORT_PATH, index=False)
    print("[OK] CSV saved:", REPORT_PATH)
    return df

# ===================== RUN =====================
if os.path.exists(DB_PATH):
    os.remove(DB_PATH)

init_db()
enroll()
build_models()
visualize_clusters()
recognize()
export_csv()
```
